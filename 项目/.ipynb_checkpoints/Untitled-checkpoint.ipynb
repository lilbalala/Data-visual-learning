# 数据导入与预处理

## 数据导入

```python
import pandas as pd
import matplotlib.pyplot as plt
from pyecharts.charts import Line,Pie,WordCloud,Bar,Grid,Page
from pyecharts import options as opts
import seaborn as sns
```

```python
mum_baby=pd.read_csv('mum_baby.csv')
mum_baby_trade_history=pd.read_csv('mum_baby_trade_history.csv')
```

## 数据查看

```python
print(mum_baby.info())
print(mum_baby_trade_history.info())
```

```python
mum_baby_trade_history.describe()
```

## 查看重复值

```python
print(mum_baby[mum_baby.duplicated()])
print(mum_baby_trade_history[mum_baby_trade_history.duplicated()])
```

## 转为日期格式

```python
mum_baby['birthday']=pd.to_datetime(mum_baby['birthday'].astype(str))
mum_baby_trade_history['day']=pd.to_datetime(mum_baby_trade_history['day'].astype(str))
```

## 异常值

```python
# 用户基本信息表中，生日当中有2010年前的，进行剔除；性别当中有未知性别，剔除未知性别数据
mum_baby_c=mum_baby[mum_baby['birthday']>='2010-01-01']
mum_baby_c=mum_baby_c[mum_baby_c['gender']!=2]
```

```python
# 交易信息表中，buy_mount最大值为10000，不符合人们的购物习惯，个人消费基本上不会出现一次购买10000件商品，从数据当中
# 也可以看出，大部分都是购买1件。一般我们取标准差(std)3倍以内的数据作为正常区间，63 * 3 = 189，我们保留buy_mount 189以内的数据
mum_baby_trade_history_c=mum_baby_trade_history[mum_baby_trade_history['buy_mount']<=189]
```

## 合并数据

```python
mum_baby_trade_history_c=pd.merge(mum_baby_trade_history_c,mum_baby_c,on='user_id',how='left')
```

## 获取交易的年、月、日和季度

```python
mum_baby_trade_history_c['t_year']=mum_baby_trade_history_c['day'].dt.year
mum_baby_trade_history_c['t_month']=mum_baby_trade_history_c['day'].dt.month
mum_baby_trade_history_c['t_day']=mum_baby_trade_history_c['day'].dt.day
mum_baby_trade_history_c['t_quarter']=mum_baby_trade_history_c['day'].dt.quarter
```

# 数据分析与可视化

## 年度销量

```python
t_year_mount=mum_baby_trade_history_c.groupby('t_year')['buy_mount'].sum()
t_year_mount
```

```python
# 绘制柱状图
plt.rcParams['font.family'] = 'SimHei'
plt.bar([str(x)+'年' for x in t_year_mount.index],t_year_mount.values)
plt.xlabel('年份')
plt.ylabel('销量')
plt.title('年份销售情况')
plt.show()
```

## 月份销量情况

```python
t_y_m_mount=mum_baby_trade_history_c.groupby(['t_year','t_month'])['buy_mount'].sum()
t_y_m_mount
```

```python
# 绘制折线图
line_one=Line()
line_one.add_xaxis([str(x) for x in t_y_m_mount.index.tolist()])
line_one.add_yaxis('',t_y_m_mount.tolist())
line_one.set_global_opts(title_opts=opts.TitleOpts(title='年月销量趋势图'))
line_one.render_notebook()
```

## 季度销量情况

```python
t_y_q_mount=mum_baby_trade_history_c.groupby(['t_year','t_quarter'])['buy_mount'].sum()
t_y_q_mount
```

```python
line_two=Line()
line_two.add_xaxis([str(x) for x in t_y_q_mount.index.tolist()])
line_two.add_yaxis('',t_y_q_mount.tolist())
line_two.set_global_opts(title_opts=opts.TitleOpts(title='年季度销量趋势图'))
line_two.render_notebook()
```

## 商品一级类目销量情况

```python
category_1_mount=mum_baby_trade_history_c.groupby('category_1')['buy_mount'].sum()
category_1_mount
```

```python
datapie=[x for x in zip(category_1_mount.index.tolist(),category_1_mount.tolist())]
pie_one=Pie()
pie_one.add('',datapie,rosetype='area',radius=['30%','50%'],label_opts=opts.LabelOpts(formatter='{b}:{c}:{d}%'))
pie_one.set_global_opts(title_opts=opts.TitleOpts(title='商品一级类目的销量情况'))
pie_one.render_notebook()
```

## 商品二级类目订单情况

```python
category_2_mount=mum_baby_trade_history_c['category_2'].value_counts()
category_2_mount
```

```python
datawc=[x for x in zip(category_2_mount.index.tolist(),category_2_mount.tolist())]
wc=WordCloud()
wc.add('',datawc,word_size_range=[10,60])
wc.set_global_opts(title_opts=opts.TitleOpts(title='商品二级类目的销量情况'))
wc.render_notebook()
```

## 一级类目下的二级类目分析

```python
category=mum_baby_trade_history_c[['category_1','category_2','buy_mount']]
```

```python
# 每个一级类目下分别有多少二级类目
category1_count=category[['category_1','category_2']].drop_duplicates()['category_1'].value_counts()
# 绘制柱状图
bar_one=Bar()
bar_one.add_xaxis([str(x) for x in category1_count.index.tolist()])
bar_one.add_yaxis('',category1_count.tolist())
bar_one.set_global_opts(title_opts=opts.TitleOpts(title='一级类目下二级类目数量'))
bar_one.render_notebook()
```

```python
# 每个一级类目下销量最好的二级类目
category_1_2=category.pivot_table('buy_mount',columns='category_1',index='category_2',aggfunc='sum')
category_1_2
```

```python
data_28=category_1_2.sort_values(by=28,ascending=False)[28][:5]
data_38=category_1_2.sort_values(by=38,ascending=False)[38][:5]
data_50008168=category_1_2.sort_values(by=50008168,ascending=False)[50008168][:5]
data_50014815=category_1_2.sort_values(by=50014815,ascending=False)[50014815][:5]
data_50022520=category_1_2.sort_values(by=50022520,ascending=False)[50022520][:5]
data_122650008=category_1_2.sort_values(by=122650008,ascending=False)[122650008][:5]
```

```python
pie_28=Pie()
pie_28.add('',[z for z in zip(data_28.index.tolist(),data_28.tolist())],
           rosetype='area',radius=['20%','30%'],center=['15%','25%'],label_opts=opts.LabelOpts(formatter='{b}:{c}'))
pie_28.set_global_opts(title_opts=opts.TitleOpts(title='28',pos_left='14%',pos_top='24%'),
                       legend_opts=opts.LegendOpts(is_show=False))

pie_38=Pie()
pie_38.add('',[z for z in zip(data_38.index.tolist(),data_38.tolist())],
           rosetype='area',radius=['20%','30%'],center=['50%','25%'],label_opts=opts.LabelOpts(formatter='{b}:{c}'))
pie_38.set_global_opts(title_opts=opts.TitleOpts(title='38',pos_left='49%',pos_top='24%'),
                       legend_opts=opts.LegendOpts(is_show=False))

pie_50008168=Pie()
pie_50008168.add('',[z for z in zip(data_50008168.index.tolist(),data_50008168.tolist())],
                 rosetype='area',radius=['20%','30%'],center=['85%','25%'],label_opts=opts.LabelOpts(formatter='{b}:{c}'))
pie_50008168.set_global_opts(title_opts=opts.TitleOpts(title='50008168',pos_left='82%',pos_top='24%'),
                             legend_opts=opts.LegendOpts(is_show=False))

pie_50014815=Pie()
pie_50014815.add('',[z for z in zip(data_50014815.index.tolist(),data_50014815.tolist())],
                 rosetype='area',radius=['20%','30%'],center=['15%','70%'],label_opts=opts.LabelOpts(formatter='{b}:{c}'))
pie_50014815.set_global_opts(title_opts=opts.TitleOpts(title='50014815',pos_left='12%',pos_top='69%'),
                             legend_opts=opts.LegendOpts(is_show=False))

pie_50022520=Pie()
pie_50022520.add('',[z for z in zip(data_50022520.index.tolist(),data_50022520.tolist())],rosetype='area',
                 radius=['20%','30%'],center=['50%','70%'],label_opts=opts.LabelOpts(formatter='{b}:{c}'))
pie_50022520.set_global_opts(title_opts=opts.TitleOpts(title='50022520',pos_left='47%',pos_top='69%'),
                             legend_opts=opts.LegendOpts(is_show=False))

pie_122650008=Pie()
pie_122650008.add('',[z for z in zip(data_122650008.index.tolist(),data_122650008.tolist())],rosetype='area',
                  radius=['20%','30%'],center=['85%','70%'],label_opts=opts.LabelOpts(formatter='{b}:{c}'))
pie_122650008.set_global_opts(title_opts=opts.TitleOpts(title='122650008',pos_left='81%',pos_top='69%'),
                              legend_opts=opts.LegendOpts(is_show=False))

grid=Grid(init_opts=opts.InitOpts(width='1500px',height='800px'))
grid.add(pie_28,grid_opts=opts.GridOpts())
grid.add(pie_38,grid_opts=opts.GridOpts())
grid.add(pie_50008168,grid_opts=opts.GridOpts())
grid.add(pie_50014815,grid_opts=opts.GridOpts())
grid.add(pie_50022520,grid_opts=opts.GridOpts())
grid.add(pie_122650008,grid_opts=opts.GridOpts())
grid.render_notebook()
```

## 不同性别购买的数据差异

```python
data_gender=mum_baby_trade_history_c[mum_baby_trade_history_c['gender'].notnull()]
gender_mount=data_gender.groupby('gender')['buy_mount'].mean()
gender_mount
```

```python
bar_two=Bar()
bar_two.add_xaxis(['男','女'])
bar_two.add_yaxis('',gender_mount.tolist(),label_opts={'is_show':False})
bar_two.set_global_opts(title_opts=opts.TitleOpts(title='不同性别购买的数据差异'))
bar_two.render_notebook()
```

## 不同性别对购买商品一级目录的数据差异

```python
gender_c1=data_gender.pivot_table('buy_mount','category_1','gender','sum')
gender_c1
```

```python
stbar_one=Bar()
stbar_one.add_xaxis(['男','女'])
for i in gender_c1.index.tolist():
    stbar_one.add_yaxis(str(i),gender_c1.loc[i].tolist(),stack=True,label_opts={'is_show':False})
stbar_one.set_global_opts(title_opts=opts.TitleOpts(title='不同性别对购买商品一级目录的数据差异'),
                         legend_opts=opts.LegendOpts(pos_right='0',orient='vertical'))
stbar_one.render_notebook()
```

## 不同性别对购买商品二级目录的数据差异

```python
gender_c2=data_gender.pivot_table('buy_mount','category_2','gender','sum')
gender_c2=gender_c2.fillna(0)
gender_c2
```

```python
stbar_two=Bar()
stbar_two.add_xaxis(['男','女'])
for i in gender_c2.index.tolist():
    stbar_two.add_yaxis(str(i),gender_c2.loc[i].tolist(),stack=True,label_opts={'is_show':False})
stbar_two.set_global_opts(title_opts=opts.TitleOpts(title='不同性别对购买商品二级目录的数据差异'),
                         legend_opts=opts.LegendOpts(pos_right='0',orient='vertical',type_='scroll'))
stbar_two.render_notebook()
```

## 购买商品时婴儿的年龄分布

```python
age_data=mum_baby_trade_history_c[mum_baby_trade_history_c['birthday'].notnull()].copy()
age_data['age']=round((age_data['day']-age_data['birthday']).dt.days/365,2)
age_data=age_data[age_data['age']>0]
age_data
```

```python
sns.displot(age_data['age'],kind='hist')
plt.show()
```
