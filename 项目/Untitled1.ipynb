```python
from pyecharts import options as opts
from pyecharts.charts import Grid, Line, Scatter,Pie
from pyecharts.faker import Faker

scatter = (
    Pie(init_opts=opts.InitOpts()
    .add('',[z for z in zip(Faker.choose(),Faker.values())],center=['20%','55%'])
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Grid-Scatter"),
        legend_opts=opts.LegendOpts(pos_left="20%"),
    )
)
line = (
    Pie()
    .add('',[z for z in zip(Faker.choose(),Faker.values())],center=['55%','55%'])
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Grid-Line", pos_right="5%"),
        legend_opts=opts.LegendOpts(pos_right="20%"),
    )
)

grid = (
    Grid(init_opts=opts.InitOpts(width='1500px'))
    .add(scatter, grid_opts=opts.GridOpts(pos_left="55%"))
    .add(line, grid_opts=opts.GridOpts(pos_right="55%"))
)
grid.render_notebook()
```
