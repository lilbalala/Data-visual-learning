```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
year_population_age = pd.read_csv('../data/us_population_by_age.csv')
plt.stackplot(year_population_age['year'],year_population_age['year_under5'],year_population_age['year5_19'],
              year_population_age['year45_64'],year_population_age['year65above'],
              labels=['5岁以下','5至19岁','20至44岁','45至64岁','65岁以上'])
plt.legend()
plt.xticks(rotation=45)
plt.title('美国人口老龄化')
plt.show()
```
