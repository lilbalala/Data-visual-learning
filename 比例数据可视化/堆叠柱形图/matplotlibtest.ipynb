```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pre_approval_rate = pd.read_csv('../data/presidential_approval_rate.csv')
x = pre_approval_rate['political_issue'].values
y1 = pre_approval_rate['support'].values
y2 = pre_approval_rate['oppose'].values
y3 = pre_approval_rate['no_opinion'].values
```

```python
plt.rcParams['font.family']=['sans-serif']
plt.rcParams['font.sans-serif']=['SimHei']
plt.figure(figsize=(15,9))
l1=plt.bar(x,y1,fc='r',width=0.3)
l2=plt.bar(x,y2,bottom=y1,fc='g',width=0.3)
l3=plt.bar(x,y3,bottom=y1+y2,fc='y',width=0.3)
plt.legend(handles=[l1,l2,l3],labels=['','',''],loc='best')
for x1,y1,x2,y2 in zip(x,y1,x,y2):
 plt.text(x1,y1,'%2.0f'%y1,ha='center',va='bottom')
 plt.text(x2,y1+y2,'%2.0f'%y2,ha='center',va='bottom')
plt.show()
```
